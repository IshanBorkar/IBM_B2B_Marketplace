## 3) Load the MIM data

```python
%cd GFK_Data
```

```python
path ='D:\\IBM_B2B_C_Onboarding\Data\GFK_Data'
MIM_files = glob.glob(os.path.join(path, "*.xlsx"))
df_MIM_files=None
for file in MIM_files:
    df = pd.read_excel(file,header=0,usecols=List2)
    if df_MIM_files is None:
        df_MIM_files=df  
    else:
        df_MIM_files=pd.concat([df_MIM_files,df],axis=0, ignore_index=True)
print("Shape of MIM files : ",df_MIM_files.shape)
```
